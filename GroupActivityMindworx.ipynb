```python
def fill_missing_voted_party(row, mode_party):
    if pd.isnull(row['VOTED']) and pd.isnull(row['POLITICAL PARTY']):
        row['VOTED'] = False
    elif pd.isnull(row['VOTED']) and pd.notnull(row['POLITICAL PARTY']):
        row['VOTED'] = True
    elif pd.isnull(row['POLITICAL PARTY']) and row['VOTED'] == True:
        row['POLITICAL PARTY'] = mode_party
    return row


mode_party = df['POLITICAL PARTY'].mode()[0]


df = df.apply(lambda row: fill_missing_voted_party(row, mode_party), axis=1)
```

Dealing with the political party column if NAN but didnt change in the previous change. Will change it now
